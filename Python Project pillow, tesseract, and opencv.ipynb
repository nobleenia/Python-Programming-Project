```python
import zipfile

from PIL import Image
import pytesseract
import cv2 as cv
import numpy as np
import pytesseract
from kraken import pageseg
import math

# loading the face detection classifier
face_cascade = cv.CascadeClassifier('readonly/haarcascade_frontalface_default.xml')

# the rest is up to you!
```

```python
def image_extr(name):
    zip_name = 'readonly/' + name
    output_file = []
    
    with zipfile.ZipFile(zip_name) as my_zip:
        zip_infos = my_zip.infolist()
        
        for item in zip_infos:
            # name
            name = item.filename
            # image
            image = Image.open(my_zip.open(name))
            # text
            image_strs = pytesseract.image_to_string(image.convert('L'))
            
            # test if "Christopher" or "Mark" are in the text
            if ("Christopher" in image_strs) or ("Mark" in image_strs):
                 # example of dictionary
                my_dictionary = {"name":name, "image":image, "text":image_strs}
                output_file.append(my_dictionary)
    return output_file
```

```python
small_images = image_extr("small_img.zip")
big_images = image_extr("images.zip")
```

```python
def extract_faces(image, scale_factor):
    gray = np.array(image.convert("L"))
    faces = face_cascade.detectMultiScale(gray, scale_factor)
    
    # if no faces are detected
    if (len(faces) == 0):
        return None
    
    # extract faces into the list imgs
    faces_images = []
    
    for x,y,w,h in faces:
        faces_images.append(image.crop((x,y,x+w,y+h)))
    
    # compute nrows and ncols
    n_column = 5
    n_row = math.ceil(len(faces) / n_column)
    
    # contact sheet
    contact_sheet = Image.new(image.mode, (550, 110 * n_row))
    x, y = (0, 0)
    
    for face in faces_images:
        face.thumbnail((110,110))
        contact_sheet.paste(face, (x,y))
        
        if x + 110 == contact_sheet.width:
            x = 0
            y += 110
        else:
            x += 110
            
    return contact_sheet
```

```python
def value_search(value, zip_name, scale_factor):
    if zip_name == "small_img.zip":
        ref_images = small_images
    else:
        ref_images = big_images
    
    for item in ref_images:
        # test if value is in the text
        if value in item["text"]:
            # print out the name of the figure
            print("Results found in file {}".format(item["name"]))
        
            # index out the faces
            image = item["image"]
            contact_sheet = extract_faces(image, scale_factor)
            if contact_sheet is not None:
                display(contact_sheet)
            else:
                print("But there were no faces in that file")
```

```python
value = "Christopher"
zip_name = "small_img.zip"

value_search(value, zip_name, scale_factor = 1.4)
```

```python
value = "Mark"
zip_name = "images.zip"

value_search(value, zip_name, scale_factor = 1.5)
```
